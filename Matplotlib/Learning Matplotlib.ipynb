## 2D Line Plot

###### Time series data is a collection of data points recorded over time, used to analyze trends and patterns.

```python
# From a pandas data frame
df = pd.read_csv("sharma-kohli.csv")
df
```

```python
#plotting RG Sharma performance using 2D Line graph
plt.plot(df["index"],df["RG Sharma"])
```

```python
#plotting virat kohli performance graph using 2D Line graph
plt.plot(df["index"],df["V Kohli"])
```

```python
# Plotting multiple plts
plt.plot(df["index"],df["RG Sharma"])
plt.plot(df["index"],df["V Kohli"])
# blue line indicate RG Sharma
# Orange line indicate V Kohli
```

```python
# Labels Title
plt.plot(df["index"],df["RG Sharma"])
plt.plot(df["index"],df["V Kohli"])
plt.title("RG Sharma Vs V Kohli Career Comparison")
#giving the column label and row label
plt.xlabel("Seasons")
plt.ylabel("Runs Scored")
```

```python
# color(hex) and line(width and style) and marker (size)
plt.plot(df["index"],df["RG Sharma"],color = "#29695d") # Green
plt.plot(df["index"],df["V Kohli"],color = "#493aa1") # Purple
plt.title("RG Sharma Vs V Kohli Career Comparison")
plt.xlabel("Seasons")
plt.ylabel("Runs Scored")
```

```python
# Line Style
plt.plot(df["index"],df["RG Sharma"],color = "#29695d", linestyle = "dashed") # there  so many line styles dashed, dotted,dashdot, and default is solid
plt.plot(df["index"],df["V Kohli"],color = "#493aa1", linestyle = "dashed") 
plt.title("RG Sharma Vs V Kohli Career Comparison")
plt.xlabel("Seasons")
plt.ylabel("Runs Scored")
```

```python
# Line width
plt.plot(df["index"],df["RG Sharma"],color = "#29695d", linewidth = 3) 
plt.plot(df["index"],df["V Kohli"],color = "#493aa1", linewidth = 2) 
plt.title("RG Sharma Vs V Kohli Career Comparison")
plt.xlabel("Seasons")
plt.ylabel("Runs Scored")
```

```python
# Marker
plt.plot(df["index"],df["RG Sharma"],color = "#29695d",marker = "D", markersize = 9) # You can also add + and other things and markersize increase the 
# size of the dot we put
plt.plot(df["index"],df["V Kohli"],color = "#493aa1",marker = ".",markersize = 10) 
plt.title("RG Sharma Vs V Kohli Career Comparison")
plt.xlabel("Seasons")
plt.ylabel("Runs Scored")
```

```python
# legend function in matplotlib use for labeling the lines
plt.plot(df["index"],df["RG Sharma"],label = "RG Sharma")
plt.plot(df["index"],df["V Kohli"], label = "Virat Kohli")
plt.title("RG Sharma Vs V Kohli Career Comparison")
plt.xlabel("Seasons")
plt.ylabel("Runs Scored")
plt.legend()
```

```python
# plt.scatter on pandas data frame
batter = pd.read_csv("batter.csv")
batter
```

```python
plt.scatter(first_50_batters["avg"],first_50_batters["strike_rate"])
plt.title("Avg and Strike rate of top 50 batmans")
plt.xlabel("Avg")
plt.ylabel("Strike rate")
```

### BAR CHART

#### . Bivariate and Univarite Analysis

#### . Numerical vs Catagorical 

#### . Use case - Aggregate Analysis of groups

```python
# Simple Bar chart
students = [10,34,65,45]
color = ["Green","Blue","Orange","Yello"]
plt.bar(color,students,color = "black")
plt.title("Color Choices")
plt.xlabel("Colors")
plt.ylabel("Students")
```

```python
# Ploting horizontal bar chart
plt.barh(color,students,color = "black") #for the horizontal barchart we use the plt.barh
```

```python
# Face Challenge of plotting Multiple Bar charts
df = pd.read_csv("batsman_season_record.csv")
df
```

```python
plt.bar(df["batsman"],df["2015"])
```

```python
#Now ploting side by side 2016 year plot
plt.bar(df["batsman"],df["2015"])
plt.bar(df["batsman"],df["2016"])
```

##### All batsman make more score than 2015 except the MS DHONI. The 2016 key has been placed on top of the 2015 key

#### Multiple bar plots

```python
# Way to plot this Multiple Bar Plots
plt.bar(np.arange(df.shape[0])-0.2,df["2015"],width = 0.2,color = "yellow")
plt.bar(np.arange(df.shape[0]),df["2016"],width = 0.2,color = "red")
plt.bar(np.arange(df.shape[0])+ 0.2,df["2017"],width = 0.2,color = "black")
# labeling the x axis
plt.xticks(np.arange(df.shape[0]),df["batsman"])
#titleing the graph
plt.title("All Players 2015,2016,2017 matches performance")
plt.show()
```

```python
# A problem
students = [10,34,65,45]
color = ["Green greengreengreen green","Blueblue blue lbue blue","Orangeorangeorganegoranger","Yello Yello Yello Yello Yello"]
plt.bar(color,students,color = "black")
```

##### The names of columns overlapped

```python
#To avoid this
students = [10,34,65,45]
color = ["Green greengreengreen green","Blueblue blue lbue blue","Orangeorangeorganegoranger","Yello Yello Yello Yello Yello"]
plt.bar(color,students,color = "black")
# to avoid the problem of overlapping columns name if they are bigger 
plt.xticks(rotation = "vertical")
```

# Stacked Bar Chart

```python
plt.bar(df["batsman"],df["2017"],label = '2017')
plt.bar(df["batsman"],df["2016"],bottom = df["2017"],label = '2016')
plt.bar(df["batsman"],df["2015"],bottom = (df["2016"] + df["2017"]),label = '2015')
plt.legend()
plt.show()
```

#### Pie Chart

```python
# simple data
data = [34,32,12,54,77]
subjects = ["pakstudy","English","Computer","Discreate math","ICT"]
plt.pie(data,labels = subjects)
plt.show()
```

```python
# pratical Data Set
df = pd.read_csv("gayle-175.csv")
df
```

```python
plt.pie(df["batsman_runs"],labels = df["batsman"],autopct = "%.1f%%")
plt.show()
```

```python
# Colors
plt.pie(df["batsman_runs"],labels = df["batsman"],autopct = "%.1f%%",colors = ["blue","green","yellow","black","red","pink"])
plt.show()
```

```python
# explode shadows
plt.pie(df["batsman_runs"],labels = df["batsman"],autopct = "%.1f%%",explode = [0.1,0,0,0,0,0],shadow = True)
plt.show()
```
